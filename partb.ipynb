```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import seaborn as sns

# William Winwood Reade’s book “The Martyrdom of Man”
# first few chapters
text = open("Hackathon.txt").read()
tokens_without_sw = text.split()
print(tokens_without_sw)
```

```python
def common(List):
    count = 0
    number = List[0]
     
    for i in List:
        count_frequency = List.count(i)
        if(count_frequency> count):
            count = count_frequency
            number = i
 
    return number
```

```python
print(common(tokens_without_sw))
```

```python
stopwords = [',', '.', 'I', '?', ';', 'the', 'and', 'for', 'of', 'to','be','in','a', 'is','by','not','it','or','as','all','no','any','can','will','with','are','have','this','which','so','you','that','from','upon','do','there','their','such','but','shall','than','they','on','may','its','if','while','who','It','If','has','these','In','at','The']
for x in stopwords:
  count4 = tokens_without_sw.count(x)
  for y in range(count4):
    tokens_without_sw.remove(x)

print(tokens_without_sw)
print(common(tokens_without_sw))
```

```python
dictword = {}
new_list = tokens_without_sw.copy()

for x in range(31):
    most_common = common(new_list)
    most_count = new_list.count(most_common)
    dictword[most_common] = most_count
    for y in range(most_count):
      while most_common in new_list: new_list.remove(most_common)


print(dictword)
  
```

```python
from pandas import DataFrame
d = {'Word':dictword.keys(), 'Count': dictword.values()}
df = DataFrame(data=d)
df.head()
```
